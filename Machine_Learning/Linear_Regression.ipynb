```python
import numpy as np
```

```python

class Linear_Regression:
    def __init__(self):
        self.weights = None

    def hyp(self, x):
        return x.dot(self.weights)

    def loss(self, x, y):
        pred = self.hyp(x)
        error = y - pred
        return np.mean(error ** 2)

    def gradient(self, x, y):
        pred = self.hyp(x)
        error = y - pred
        return -2 * np.mean(x.T.dot(error))
    
    def gradient_descent(self, x, y, lr, epochs):
        self.weights = np.zeros((x.shape[1]))

        for e in range(epochs):
            self.weights -= lr * self.gradient(x, y)

            if e % 10 == 0:
                print(f"Epoch {e}/{epochs} - Loss : {self.loss(x, y)}")

    def train(self, x, y, lr, epochs):
        x = np.hstack((x, np.ones((x.shape[0], 1))))
        self.gradient_descent(x, y, lr, epochs)

    def predict(self, x):
        x = np.hstack((x, np.ones((x.shape[0], 1))))
        return self.hyp(x)


if __name__ == '__main__':

    x_train = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])
    y_train = np.array([1, 2, 3, 4])

    x_test = np.array([[9, 10], [11, 12]])

    model = Linear_Regression()
    model.train(x_train, y_train, 0.001, 100)
    print(model.predict(x_test))
```
