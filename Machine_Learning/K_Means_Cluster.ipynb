```python
import numpy as np
import random
```

```python

class KMeansCluster:

    def __init__(self, k, max_iters):
        self.k = k
        self.max_iters = max_iters

    def predict(self, x):
        
        centroids = random.sample(list(x), k=self.k)
        clusters = []

        for i in range(self.max_iters):
            for point in x:
                distance = np.linalg.norm( centroids - point, axis=1)
                clusters.append(np.argmin(distance))

            clusters = np.array(clusters)
            new_centroids = []
            for i in range(len(centroids)):
                cluster = x[np.where(clusters == i)[0]]
                new_centroids.append(np.mean(cluster, axis=0))
            
            centroids = new_centroids

        return clusters



if __name__ == '__main__':

    model = KMeansCluster(3, 1)

    X_train = np.array([
        [1.0, 2.0],   # Cluster 0
        [1.5, 1.8],   # Cluster 0
        [2.0, 2.5],   # Cluster 0
        [8.0, 9.0],   # Cluster 1
        [8.5, 8.0],   # Cluster 1
        [9.0, 8.5],   # Cluster 1
        [5.0, 5.0],   # Cluster 2
        [5.5, 4.8],   # Cluster 2
        [4.5, 5.2]    # Cluster 2
    ])
    X_test = np.array([
        [1.2, 2.2],   # Should belong to Cluster 0
        [8.8, 8.2],   # Should belong to Cluster 1
        [5.2, 5.1],   # Should belong to Cluster 2
        [2.1, 2.3],   # Should belong to Cluster 0
        [9.1, 8.3]    # Should belong to Cluster 1
    ])
    y_test_correct = np.array([0, 1, 2, 0, 1])  # Expected cluster labels for X_test

    model.predict(X_train)
    pred = model.predict(X_test)

    print("\n\nTraining Data (X_train):\n", X_train)
    print("\nTesting Data (X_test):\n", X_test)
    print("\nCorrect Cluster Assignments for X_test (y_test_correct):\n", y_test_correct)
    print(f'\nPred : {pred}')

```
