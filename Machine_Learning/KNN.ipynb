```python
import numpy as np
```

```python


class KNN:
    def __init__(self):
        self.inputs = None
        self.labels = None
        self.k = None

    def train(self, inputs, labels, k):
        self.inputs = inputs
        self.labels = labels
        self.k = k        

    def predict(self, samples):

        results = []

        for sample in samples:
            distance = np.linalg.norm( self.inputs - sample, axis=1)
            sorted_distance = np.argsort(distance)
            k_nearest_distance = self.labels[sorted_distance[:self.k]]
            _, counts = np.unique(k_nearest_distance, return_counts=True)
            results.append(k_nearest_distance[np.argmax(counts)])

        return results
        

if __name__ == '__main__':

    X_train = np.array([
        [1.0, 2.0],
        [1.5, 1.8],
        [2.0, 2.5],  # These first three points are grouped together (Class 0)
        [8.0, 9.0],
        [8.5, 8.0],
        [9.0, 8.5]   # These last three points are grouped together (Class 1)
    ])
    y_train = np.array([0, 0, 0, 1, 1, 1])
    X_test = np.array([
        [1.2, 2.2],  # Close to points of Class 0
        [8.8, 8.2],  # Close to points of Class 1
        [5.0, 5.0]   # Somewhere in the middle
    ])

    y_test = np.array([0, 0, 1])

    model = KNN()
    model.train(X_train, y_train, 3)
    predictions = model.predict(X_test)

    print("\nTraining Features (X_train):\n", X_train)
    print("\nTraining Labels (y_train):", y_train)
    print("\nTesting Features (X_test):\n", X_test)
    print("\nTesting Labels (y_test):", y_test)
    print("\nPredictions:", predictions)
```
